#This program converts image to pencil sketch

```python
from google.colab.patches import cv2_imshow

```

```python
import cv2  
```

```python
#img=cv2.imread("/content/peacock.jpg")
img=cv2.imread("/content/TAYLORsWIFT2.jpg")

img
```

```python
cv2_imshow(img)
```

###1.Convert the RGB image to a grayscale image

```python
img_gray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
img_gray
```

```python
cv2_imshow(img_gray)
```

### **2**.Convert grayscle image to a negtive image (Inverting the grayscale image)

```python
img_invert=cv2.bitwise_not(img_gray)  #if your pixel value is 2 you are going to say 255 -2 so it becomes 253...ypu use bitwise not to get the pixel value inverted
img_invert
```

```python
cv2_imshow(img_invert)   
```

### 3.Applying GAUSSIAN BLUR to blur the image

```python
img_blur=cv2.GaussianBlur(img_invert,(51,51),sigmaX=5,sigmaY=5)
img_blur
```

```python
cv2_imshow(img_blur)
```

```python
img_invert = 255 - img_blur
img_invert
```

```python
final_img = cv2.divide(img_gray, img_invert, scale=256.0)
cv2_imshow(final_img)
cv2.waitKey(0)
```

```python
cv2_imshow(img)
cv2_imshow(final_img)
cv2.waitKey(0)
```
